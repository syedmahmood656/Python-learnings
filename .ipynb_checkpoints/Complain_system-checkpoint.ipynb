```python
def complain(uname,dept):
    if dept == '1':
        print(uname +" this is muncipal dept how can i help you..?")
        c = input(uname +" reply with your complain or suggestions")
        print("your response has been recorded as " + c + ", thanks for reaching us, we will be solving on your query shortly..!")

    if dept == '2':
        print(uname +" this is educational dept how can i help you..?")
        c = input(uname +" reply with your complain or suggestions")
        print("your response has been recorded as " + c + ", thanks for reaching us, we will be solving on your query shortly..!")

    if dept == '3':
        print(uname +" this is environmental dept how can i help you..?")
        c = input(uname +" reply with your complain or suggestions")
        print("your response has been recorded as " + c + ", thanks for reaching us, we will be solving on your query shortly..!")

    if dept == '4':
        print(uname +" this is crime dept how can i help you..?")
        c = input(uname +" reply with your complain or suggestions")
        print("your response has been recorded as " + c + ", thanks for reaching us, we will be solving on your query shortly..!")

    else:
        print("dept not found, try again")
```

```python
def welcome():
    print('welcome, please enter your username..!')
    uname = input("User_name:")
    print("for muncipal dept : 1 \nfor education dept : 2 \nfor environmental : 3 \nfor criminal dept : 4")
    dept = input("enter the dept name: ")
    complain(uname,dept)
```
