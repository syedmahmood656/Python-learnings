```python
import csv
import os
```

```python
file_name = "students.csv"
```

```python
def add_std():
    print('0')
    sid = input("enter student ID")
    sname = input("enter student name")
    m_marks = input("enter maths marks : ")
    s_marks = input("enter science marks : ")
    e_marks = input("enter english marks : ")

    with open(file_name, "a",newline='') as file:
        writer =csv.write(file)
        writer.writerow([sid, sname, m_marks,s_marks,e_marks])
    print("student record recorded sucessfully")

def display_std():
    print('1')

def search_std():
    print('2')

def calculate_result():
    print('3')

def delete_std():
    print('4')

def update_std():
    print('5')
```

```python
def main_menue():

    while True:

        print("STUDENT RESULT MANAGEMENT SYSTEM")
        s = input("enter a number")

        if s == '1':
            add_std()
        elif s == '2':
            display_std()
        elif s == '3':
            search_std()
        elif s == '4':
            calculate_result()
        elif s == '5':
            delete_std()
        elif s == '6':
            update_std()
        else:
            print("invaid choice")
```

```python
main_menue()
```
